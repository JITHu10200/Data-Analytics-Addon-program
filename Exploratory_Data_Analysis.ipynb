```python
plt.figure(figsize=(8,5))
plt.scatter(data["Survived"],data["Fare"],c=data["Survived"])
plt.title("scatterplot:Survived vs Fare",fontsize=14)
plt.xlabel("Survived")
plt.ylabel("Fare")
plt.colorbar(label="Survived")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(data["Survived"],bins=30,color='red',endcolor="black")
plt.title("Histogram: Survived", fontsize=14)
plt.xlabel("Survived")
plt.ylabel("Frequency")
plt.show()
```
