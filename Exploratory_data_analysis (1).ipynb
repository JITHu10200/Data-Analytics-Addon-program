```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data.drop('Embarked',axis=1,inplace=True)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(data["Age"],data["Fare"],c=data["Survived"])
plt.title("scatterplot:Age vs Fare",fontsize=14)
plt.xlabel("Age")
plt.ylabel("Fare")
plt.colorbar(label="Survived")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(data["Age"],bins=30,color='red',endcolor='black')
plt.title("Histogram: Age",fontsize=14)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```
